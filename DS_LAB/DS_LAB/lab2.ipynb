```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
df=pd.read_csv('auto-mpg.csv')
```

```python
df
```

```python
df.isna().sum()
```

```python
df['displacement']=df['displacement'].fillna(df['displacement'].mean())
```

```python
df['horsepower']=df['horsepower'].fillna(df['horsepower'].mean())
```

```python
df.isna().sum()
```

```python
df
```

```python
c=[['mpg','displacement','horsepower','acceleration']]
```

```python
for i in c:
    
    sb.histplot(df[i])
    plt.show()
```

```python
sb.scatterplot(x='acceleration' , y='mpg',data=df)
plt.show()
```

```python
sb.countplot(x=df['acceleration'])
```

```python
sb.pointplot(x=df['cylinders'] , y=df['mpg'] ,data=df)
```
