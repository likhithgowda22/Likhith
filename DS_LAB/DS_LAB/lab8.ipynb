```python
nltk.download('stopwords')
from nltk.corpus import stopwords
```

```python
stopwords_list=stopwords.words('english')
```

```python
stopwords_list
```

```python
plt.figure(figsize=(10,10))
```

```python
wc=WordCloud(height=500,width=500).generate(data)
```

```python
plt.imshow(wc)
```

```python
from nltk import word_tokenize
```

```python
nltk_token=nltk.word_tokenize(str(data))
```

```python
nltk_token
```

```python
words=[]
```

```python
for word in nltk_token:
    if word.isalpha():
        words.append(word)
    words
```

```python
words
```

```python
clean_words=[]
```

```python
for word in words:
    if word not in stopwords_list:
        clean_words.append(word)
    clean_words
```

```python
clean_words
```

```python
type(clean_words)
```

```python
newwords=" ".join(clean_words)
```

```python
type(newwords)
```

```python
plt.figure(figsize=(10,10))
```

```python
wc=WordCloud().generate(newwords)
```

```python
plt.imshow(wc)
```

```python
wc=WordCloud(width=500 , height=500,background_color='white',colormap='Reds').generate(newwords)
```

```python
plt.imshow(wc)
plt.show()
```

```python
plt.axis('off')
plt.imshow(wc)
plt.show()
```
